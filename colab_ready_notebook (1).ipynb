```python
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split

from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import ConvertToNumpyArray
```

```python

# Standard imports
import json, os, math, random, time
import numpy as np
from google.colab import files
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split

from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import ConvertToNumpyArray

from sklearn.metrics import mean_squared_error, r2_score

# set seed for reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
```

```python
# Option A: Upload files directly (small/easy). Choose affinity.txt, ligands.txt, proteins.txt
print("Upload affinity.txt, ligands.txt, proteins.txt (or mount Drive below).")
uploaded = files.upload()  # use the file-picker

# Option B: (commented) Mount Google Drive if files are stored in Drive.
# from google.colab import drive
# drive.mount('/content/drive')
# Then set paths like: ligands_path = "/content/drive/MyDrive/path/ligands.txt"
```

```python
# Adjust filenames if your uploaded files have slightly different names
ligands_path = [k for k in uploaded.keys() if "ligand" in k.lower()][0]
proteins_path = [k for k in uploaded.keys() if "protein" in k.lower()][0]
affinity_path = [k for k in uploaded.keys() if "affinity" in k.lower()][0]

print("Using:", ligands_path, proteins_path, affinity_path)

# ligands and proteins expected as JSON dictionaries: {"id": "SMILES", ...}, {"id": "SEQ", ...}
with open(ligands_path, "r") as f:
    ligands = json.load(f)

with open(proteins_path, "r") as f:
    proteins = json.load(f)

# affinity.txt: numeric matrix where each ROW corresponds to ligand (same order as ligands.keys())
# and each COLUMN corresponds to protein (same order as proteins.keys()).
aff = np.loadtxt(affinity_path)   # shape (n_ligands, n_proteins)
aff = np.nan_to_num(aff, nan=0.0)  # guard
print("Parsed sizes -> ligands:", len(ligands), "proteins:", len(proteins), "affinity:", aff.shape)
```

```python
# Ligand: Morgan fingerprint -> numpy vector of 0/1 length n_bits
N_BITS = 1024

def smiles_to_fp_array(smiles, n_bits=N_BITS):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return np.zeros(n_bits, dtype=np.float32)
        fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=n_bits)
        arr = np.zeros((n_bits,), dtype=np.int8)
        ConvertToNumpyArray(fp, arr)  # writes 0/1 ints into arr
        return arr.astype(np.float32)
    except Exception as e:
        # fallback zero vector on any parse error
        return np.zeros(n_bits, dtype=np.float32)

# Protein: amino-acid composition vector (20 aa frequencies) + normalized length -> 21-d
AMINO_ACIDS = "ACDEFGHIKLMNPQRSTVWY"
AA_IDX = {a:i for i,a in enumerate(AMINO_ACIDS)}

def protein_to_aac_vector(seq):
    seq = seq.upper()
    counts = np.zeros(len(AMINO_ACIDS), dtype=np.float32)
    total = 0
    for ch in seq:
        if ch in AA_IDX:
            counts[AA_IDX[ch]] += 1
            total += 1
    if total > 0:
        freqs = counts / total
    else:
        freqs = counts
    # additional normalized length feature (log scaled)
    length_feature = np.array([math.log1p(len(seq)) / 10.0], dtype=np.float32)
    return np.concatenate([freqs, length_feature])  # shape (21,)
```

```python
# Keep a stable ordering for mapping to affinity matrix:
ligand_ids = list(ligands.keys())
protein_ids = list(proteins.keys())

print("Converting ligands to fingerprints...")
X_drug = np.stack([smiles_to_fp_array(ligands[lid]) for lid in tqdm(ligand_ids)])
print("X_drug shape:", X_drug.shape)

print("Converting proteins to AAC vectors...")
X_prot = np.stack([protein_to_aac_vector(proteins[pid]) for pid in tqdm(protein_ids)])
print("X_prot shape:", X_prot.shape)

# Check affinity dimensions match ordering
assert aff.shape[0] == X_drug.shape[0], "affinity rows != #ligands"
assert aff.shape[1] == X_prot.shape[0],    "#affinity cols != #proteins"

# Save features for reuse
np.save("X_drug_base.npy", X_drug)
np.save("X_prot_base.npy", X_prot)
np.save("affinity_matrix.npy", aff)
print("Saved base feature files: X_drug_base.npy, X_prot_base.npy, affinity_matrix.npy")
```

```python
# BUILD PAIRS: each sample = (drug_i, prot_j) with label aff[i,j] (scalar)
# NOTE: full pairs = n_ligands * n_proteins; can be large. Use `max_pairs` to sample a subset.

n_L, n_P = X_drug.shape[0], X_prot.shape[0]
n_pairs = n_L * n_P
print("Total possible pairs:", n_pairs)

# If you want to limit memory/training size, set max_pairs (e.g., 200000). Set None to use all pairs.
max_pairs = None   # <- set to e.g. 200000 to randomly sample 200k pairs

if (max_pairs is None) or (n_pairs <= max_pairs):
    # use all pairs
    Xd_pairs = np.repeat(X_drug, n_P, axis=0)      # shape (n_pairs, drug_dim)
    Xp_pairs = np.tile(X_prot, (n_L, 1))           # shape (n_pairs, prot_dim)
    y_pairs = aff.reshape(-1).astype(np.float32)   # shape (n_pairs,)
else:
    # random sampled pairs
    idxs = np.random.choice(n_pairs, size=max_pairs, replace=False)
    rows = idxs // n_P
    cols = idxs % n_P
    Xd_pairs = X_drug[rows]
    Xp_pairs = X_prot[cols]
    y_pairs = aff[rows, cols].astype(np.float32)

print("Final pairs used:", Xd_pairs.shape[0])
np.save("X_d_pairs.npy", Xd_pairs)
np.save("X_p_pairs.npy", Xp_pairs)
np.save("y_pairs.npy", y_pairs)
print("Saved pairwise arrays: X_d_pairs.npy, X_p_pairs.npy, y_pairs.npy")
```

```python
# Simple PyTorch dataset for pairwise data
class PairDTADataset(Dataset):
    def __init__(self, Xd, Xp, y):
        assert Xd.shape[0] == Xp.shape[0] == y.shape[0]
        self.Xd = Xd
        self.Xp = Xp
        self.y = y

    def __len__(self):
        return self.y.shape[0]

    def __getitem__(self, idx):
        return (
            torch.tensor(self.Xd[idx], dtype=torch.float32),
            torch.tensor(self.Xp[idx], dtype=torch.float32),
            torch.tensor(self.y[idx], dtype=torch.float32).unsqueeze(0)  # shape (1,)
        )

# Load the saved pairwise arrays
Xd = np.load("X_d_pairs.npy")
Xp = np.load("X_p_pairs.npy")
y = np.load("y_pairs.npy")

dataset = PairDTADataset(Xd, Xp, y)
print("Dataset size:", len(dataset))

# Train/test split
train_frac = 0.8
train_size = int(train_frac * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

batch_size = 64
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
test_loader  = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

print("Train batches:", len(train_loader), "Test batches:", len(test_loader))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

drug_dim = Xd.shape[1]   # 1024
prot_dim = Xp.shape[1]   # 21
hidden_dim = 256

class DTA_MLP(nn.Module):
    def __init__(self, drug_dim, prot_dim, hidden_dim=256):
        super().__init__()
        self.drug_net = nn.Sequential(
            nn.Linear(drug_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, hidden_dim),
            nn.ReLU()
        )
        self.prot_net = nn.Sequential(
            nn.Linear(prot_dim, 128),
            nn.ReLU(),
            nn.Linear(128, hidden_dim),
            nn.ReLU()
        )
        self.head = nn.Sequential(
            nn.Linear(hidden_dim*2, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, 1)  # scalar affinity per pair
        )

    def forward(self, d, p):
        d_feat = self.drug_net(d)
        p_feat = self.prot_net(p)
        x = torch.cat([d_feat, p_feat], dim=1)
        return self.head(x).squeeze(1)  # returns shape (batch,)

model = DTA_MLP(drug_dim, prot_dim, hidden_dim).to(device)
print(model)
```

```python
# Training hyperparams
max_epochs = 50     # <= 50 as requested
lr = 1e-3
weight_decay = 1e-5

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)

best_val_mse = float("inf")
save_path = "dta_model.pth"

for epoch in range(1, max_epochs+1):
    model.train()
    running_loss = 0.0
    count = 0
    for d_batch, p_batch, y_batch in train_loader:
        d_batch = d_batch.to(device)
        p_batch = p_batch.to(device)
        y_batch = y_batch.to(device).squeeze(1)  # shape (batch,)

        optimizer.zero_grad()
        outputs = model(d_batch, p_batch)  # shape (batch,)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * d_batch.size(0)
        count += d_batch.size(0)

    train_mse = running_loss / count

    # validation
    model.eval()
    all_preds, all_trues = [], []
    with torch.no_grad():
        for d_batch, p_batch, y_batch in test_loader:
            d_batch = d_batch.to(device)
            p_batch = p_batch.to(device)
            outputs = model(d_batch, p_batch).cpu().numpy()
            all_preds.append(outputs)
            all_trues.append(y_batch.numpy().squeeze(1))
    all_preds = np.concatenate(all_preds)
    all_trues = np.concatenate(all_trues)
    val_mse = mean_squared_error(all_trues, all_preds)
    val_r2 = r2_score(all_trues, all_preds)

    print(f"Epoch {epoch:02d}/{max_epochs} — train MSE: {train_mse:.4f}  val MSE: {val_mse:.4f}  R2: {val_r2:.4f}")

    # save best
    if val_mse < best_val_mse:
        best_val_mse = val_mse
        torch.save({
            "model_state": model.state_dict(),
            "drug_dim": drug_dim,
            "prot_dim": prot_dim,
            "hidden_dim": hidden_dim
        }, save_path)
        print("  -> saved best model.")
```

```python
# load best model
ckpt = torch.load(save_path, map_location=device)
model.load_state_dict(ckpt["model_state"])
model.to(device)
model.eval()

# compute test metrics
preds, trues = [], []
with torch.no_grad():
    for d_batch, p_batch, y_batch in test_loader:
        d_batch = d_batch.to(device)
        p_batch = p_batch.to(device)
        out = model(d_batch, p_batch).cpu().numpy()
        preds.append(out)
        trues.append(y_batch.numpy().squeeze(1))
preds = np.concatenate(preds)
trues = np.concatenate(trues)
print("Test MSE:", mean_squared_error(trues, preds))
print("Test R2 :", r2_score(trues, preds))
```

```python
# Reusable encode + predict functions

def encode_smiles(smiles):
    return torch.tensor(smiles_to_fp_array(smiles).astype(np.float32)).unsqueeze(0).to(device)  # shape (1,1024)

def encode_protein(seq):
    return torch.tensor(protein_to_aac_vector(seq).astype(np.float32)).unsqueeze(0).to(device)  # shape (1,21)

def predict_pair(smiles, seq, model=model):
    model.eval()
    d = encode_smiles(smiles)
    p = encode_protein(seq)
    with torch.no_grad():
        out = model(d, p).cpu().item()
    return float(out)

# Example interactive usage:
user_smiles = input("Enter SMILES string (or press enter to skip): ").strip()
user_seq = input("Enter protein FASTA/sequence (or press enter to skip): ").strip()
if user_smiles and user_seq:
    pred_val = predict_pair(user_smiles, user_seq)
    print("Predicted affinity:", pred_val)
else:
    print("No pair provided — use predict_pair(smiles, seq) function.")
```
